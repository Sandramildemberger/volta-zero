```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sinasc = pd.read_csv('sinasc_RO_2019.csv')

sinasc.head()
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='IDADEMAE', data=sinasc)
plt.title('Boxplot da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='PESO', data=sinasc)
plt.title('Boxplot do Peso do Bebê')
plt.xlabel('Peso do Bebê (g)')
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
plt.figure(figsize=(10, 6))
sns.displot(
    data=sinasc,
    x='IDADEMAE',
    bins=20,
    element='step',
    kde=True,
    alpha=0.25
)
plt.title('Histograma da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.ylabel('Frequência')

plt.xticks(rotation=30)

plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
plt.figure(figsize=(10, 6))
sns.displot(
    data=sinasc,
    x='PESO',
    bins=20,
    element='step',
    kde=True,
    alpha=0.25,
    color='blue'  # cor das barras
)

# Para personalizar a cor da linha do KDE em vermelho
kde_line = sns.kdeplot(
    data=sinasc['PESO'],
    color='red',  # cor da linha do KDE
    lw=2  # largura da linha do KDE
)

plt.title('Histograma do Peso do Bebê')
plt.xlabel('Peso do Bebê (gramas)')
plt.ylabel('Frequência')

# Rotacionar os rótulos do eixo X (se necessário)
plt.xticks(rotation=30)

# Mostrar o gráfico
plt.show()
```
