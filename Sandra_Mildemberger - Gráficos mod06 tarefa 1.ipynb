#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sinasc = pd.read_csv('sinasc_RO_2019.csv')

sinasc.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Gráfico para APGAR1
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='APGAR1', data=sinasc)  
plt.title('Distribuição de APGAR1')
plt.xlabel('APGAR1')
plt.ylabel('Contagem')
ax.tick_params(axis='x', rotation=30)  # Rotaciona os rótulos do eixo X em 30 graus
plt.show()

# Gráfico para APGAR5
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='APGAR5', data=sinasc) 
plt.title('Distribuição de APGAR5')
plt.xlabel('APGAR5')
plt.ylabel('Contagem')
ax.tick_params(axis='x', rotation=30)  # Rotaciona os rótulos do eixo X em 30 graus
plt.show()

# Gráfico para SEXO
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='SEXO', data=sinasc)  
plt.title('Distribuição por SEXO')
plt.xlabel('SEXO')
plt.ylabel('Contagem')
ax.tick_params(axis='x', rotation=30)  # Rotaciona os rótulos do eixo X em 30 graus
plt.show()

# Gráfico para GESTACAO
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='GESTACAO', data=sinasc)  
plt.title('Distribuição de GESTACAO')
plt.xlabel('GESTACAO')
plt.ylabel('Contagem')
ax.tick_params(axis='x', rotation=30)  
plt.show()

plt.figure(figsize=(10, 6))
ax = sns.countplot(x='ESTCIVMAE', data=sinasc)  
plt.title('Distribuição de ESTCIVMAE')
plt.xlabel('ESTCIVMAE')
plt.ylabel('Contagem')
ax.tick_params(axis='x', rotation=30)  
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='CONSULTAS', y='APGAR5', data=sinasc)
plt.title('APGAR5 Médio por CONSULTAS')
plt.xlabel('CONSULTAS')
plt.ylabel('APGAR5 Médio')
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(2, 2, figsize=(15, 12))

sns.countplot(ax=axes[0, 0], x='SEXO', data=sinasc)
axes[0, 0].set_title('Distribuição por SEXO')
axes[0, 0].tick_params(axis='x', rotation=30) 

sns.countplot(ax=axes[0, 1], x='GESTACAO', data=sinasc)
axes[0, 1].set_title('Distribuição de GESTACAO')
axes[0, 1].tick_params(axis='x', rotation=30)  

sns.countplot(ax=axes[1, 0], x='ESTCIVMAE', data=sinasc)
axes[1, 0].set_title('Distribuição de ESTCIVMAE')
axes[1, 0].tick_params(axis='x', rotation=30)  

sns.countplot(ax=axes[1, 1], x='CONSULTAS', data=sinasc)
axes[1, 1].set_title('Distribuição de CONSULTAS')
axes[1, 1].tick_params(axis='x', rotation=30)  

plt.tight_layout()
plt.show()

```
